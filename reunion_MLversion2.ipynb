# Data Preprocessing

```python
# dropping these columns due to high number of missing values.

data = data.drop(columns=['Telephone', 'Other_EMI_plans', 'Balance_in_existing_bank_account_(lower_limit_of_bucket)', 'Balance_in_existing_bank_account_(upper_limit_of_bucket)'])

# id carry no meaning and information
data = data.drop(columns = ['loan_application_id', 'applicant_id'])
```

```python
# Function to extract numeric values from strings
def extract_years(data):
    return data.str.extract('(\d+)').astype(float)

# Apply the function to the relevant columns
data['Has_been_employed_for_at_least'] = extract_years(data['Has_been_employed_for_at_least'])
data['Has_been_employed_for_at_most'] = extract_years(data['Has_been_employed_for_at_most'])
```

```python
(data['Has_been_employed_for_at_least'].min(), data['Has_been_employed_for_at_least'].min())
```

```python
# Fill missing values these columns with 0, so to ensure the we don't label high risk applicant as low risk.
data['Has_been_employed_for_at_least'] = data['Has_been_employed_for_at_least'].fillna(0)
data['Has_been_employed_for_at_most'] = data['Has_been_employed_for_at_most'].fillna(0)
```

```python
# we fill the missing values with the most freq values for high risk == 1, so to ensure the we don't label high risk applicant as low risk.
data[data['high_risk_applicant']==1][['Purpose', 'Property', 'Savings_account_balance']].describe()
```

```python
data[['Purpose', 'Property', 'Savings_account_balance']] = data[['Purpose', 'Property', 'Savings_account_balance']].fillna(data[['Purpose', 'Property', 'Savings_account_balance']].mode().iloc[0])
```

```python
# Verify there are no more missing values
data.isnull().sum()
```

##Customer Segmentation Based on Risk Segments Based on Risk:

**High Risk:** Applicants who are likely to default.
Characteristics:
* Poor credit history
* high loan amount relative to income
* short employment duration
* high number of existing loans.

**Low Risk:** Applicants who are likely to repay the loan.
Characteristics:
* Good credit history
* stable employment
* reasonable loan amount relative to income
* fewer existing loans.

##Approval Recommendations for Segments Proposed Segments for Approval:

Low-Risk Applicants:

* Good credit history (e.g., 'All Paid', 'Existing Paid').
* Stable employment (longer employment duration).
* Moderate loan amount relative to income.
* Fewer existing loans.

High-Risk Applicants:

* Applicants with 'Critical' credit history should be scrutinized more closely.
* Young applicants with stable jobs and lower loan amounts might still be considered for approval.
* Applicants with more credit accounts need thorough assessment but could be creditworthy if they have a good repayment history.

# ML models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer


# Define features and target
X = data.drop(columns=['high_risk_applicant'])
y = data['high_risk_applicant']

# List of categorical features and numerical features
categorical_features = cat_features
numerical_features = num_features

# Preprocessing pipeline for numerical features
numerical_pipeline = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# Preprocessing pipeline for categorical features
categorical_pipeline = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine pipelines into a preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, numerical_features),
        ('cat', categorical_pipeline, categorical_features)
    ]
)

# Apply preprocessing to the features
X_preprocessed = preprocessor.fit_transform(X)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.2, random_state=42)

# Output shapes for confirmation
print(f"Training set shape: {X_train.shape}")
print(f"Testing set shape: {X_test.shape}")
print(f"Training target shape: {y_train.shape}")
print(f"Testing target shape: {y_test.shape}")

```

###a. Logistic Regression
Hyperparameter Tuning:

GridSearchCV can be used to find the optimal regularization parameter (C)

```python
# Define the model and parameters
log_reg = LogisticRegression(max_iter=1000)
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100]}

# Grid Search
grid_search = GridSearchCV(log_reg, param_grid, scoring='f1', cv=5)
grid_search.fit(X_train, y_train)

# Best model
best_log_reg = grid_search.best_estimator_

# Predictions
y_pred_log_reg = best_log_reg.predict(X_test)

# Confusion Matrix
conf_matrix_log_reg = confusion_matrix(y_test, y_pred_log_reg)
print("Logistic Regression Confusion Matrix:")
print(conf_matrix_log_reg)
print(classification_report(y_test, y_pred_log_reg))
```

###b. Random Forest
Hyperparameter Tuning:

GridSearchCV for tuning parameters such as n_estimators, max_depth, and min_samples_split.

```python
from sklearn.ensemble import RandomForestClassifier

# Define the model and parameters
rf = RandomForestClassifier()
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

# Grid Search
grid_search_rf = GridSearchCV(rf, param_grid_rf, scoring='f1', cv=5)
grid_search_rf.fit(X_train, y_train)

# Best model
best_rf = grid_search_rf.best_estimator_

# Predictions
y_pred_rf = best_rf.predict(X_test)

# Confusion Matrix
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
print("Random Forest Confusion Matrix:")
print(conf_matrix_rf)
print(classification_report(y_test, y_pred_rf))
```
